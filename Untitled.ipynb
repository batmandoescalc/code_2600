```python
import pandas as pd
import numpy as np

# Load Credit data
Credit = pd.read_csv('data/Credit.csv')

# Display first 5 rows
Credit.head()

```

```python
# List feature names
Credit.columns.tolist()
```

```python
# Show data types
Credit.dtypes
```

```python
# Mean, median, mode of Age
mean_age = Credit['Age'].mean()
median_age = Credit['Age'].median()
mode_age = Credit['Age'].mode()[0]

mean_age, median_age, mode_age
```

```python
# Create subset
Subset = Credit[['Age', 'Region', 'Income']]

# Show first 5 rows
Subset.head()
```

```python
# Younger (<50) and Older (>=50)
Younger = Subset[Subset['Age'] < 50]
Older = Subset[Subset['Age'] >= 50]

len(Younger), len(Older)
```

```python
# Mean and median for each group
younger_mean = Younger['Income'].mean()
older_mean = Older['Income'].mean()
younger_median = Younger['Income'].median()
older_median = Older['Income'].median()

younger_mean, older_mean, younger_median, older_median
```

```python
# Observation about distribution
"""
If median less than mean, distribution is right-skewed.
If median is greater than mean, distribution is left-skewed.
"""
```

```python
# Unique values
Credit['Region'].unique()
```

```python
# Count of each region
Credit['Region'].value_counts()
```

```python
# Convert Student to boolean for clarity
# (optional, could just use 'Yes'/'No')
avg_income_students = Credit[Credit['Student'] == 'Yes']['Income'].mean()
avg_income_non_students = Credit[Credit['Student'] == 'No']['Income'].mean()

avg_income_students, avg_income_non_students
```

```python
# Utilization = Balance / Limit
Credit['Utilization'] = Credit['Balance'] / Credit['Limit']

# Show first 5 rows of Balance, Limit, Utilization
Credit[['Balance', 'Limit', 'Utilization']].head()
```

```python
# Maximum utilization value
Credit['Utilization'].max()
```
